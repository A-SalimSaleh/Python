```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_hospitals_in_Kenya" 

page = requests.get(url)

Soup = BeautifulSoup(page.text, 'html')
```

```python
print(Soup)
```

```python
Soup.find('table')
```

```python
Soup.find_all('table')
```

```python
Soup.find_all('table')[1]
```

```python
Soup.find('table', class_= 'wikitable sortable')
```

```python
table = Soup.find_all('table')[1]
print(table)
```

```python
Nairobi_Hospitals = table.find_all('th')
Nairobi_Hospitals
```

```python
Nairobi_Health_Hospitals = [title.text.strip() for title in Nairobi_Hospitals]
print(Nairobi_Health_Hospitals)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = Nairobi_Health_Hospitals)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    
```

```python

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
